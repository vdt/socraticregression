```python
fig, axs = plt.subplots(4, 1, figsize=(4,9), dpi=160)
plt.tight_layout(h_pad=3)

fig.suptitle('James-Stein Shrinkage', size=20, y=1.14)
radii = 2
for ax_id, shift in enumerate([1, 3, 5, 10]):
    X, Y, afterX, afterY, coloring, before_sse, after_sse = calc_shrinkage_ran(shift, radii)
    subplotter(X, Y, afterX, afterY, coloring, before_sse, after_sse, shift, axs[ax_id])
fig.text(x=0, y=1.09, s='Grey Points: Before shrinkage', color='grey', size=10)
fig.text(x=0, y=1.07, s='Blue Points: Sum of square error (SSE) from X decreased', color='b', size=10)
fig.text(x=0, y=1.05, s='Yellow Points: SSEs roughly unchanged', color='y', size=10)
fig.text(x=0, y=1.03, s='Red Points: SSEs increased', color='r', size=10)
plt.show()
```

### Commentary
- Shrinkage helps when the real value (marked by X) is close to zero. This makes sense since we're shrinking toward 0, so most points are helped toward X.
- As X gets further away from zero, James-Stein starts doing worse because the proportion of points which are helped vs hurt approaches 50-50 asymptotically.
- One other crucial detail is that at a point-wise level, James-Stein is actually a net loss because for each point that is helped, it's partner on the opposite side of X is hurt more (see below for example). The good news is that asymptotically, this weakness goes away.
- In 1 dimensions, where the helped-hurt ratio is 50-50, this detail means that James-Stein is always worse than the vanilla estimator. In 2 dimensions, there are more helped than hurt points, but it is insufficient to overcome the point-wise weakness. You can see in the last diagram that we are able to shift X far enough away that the SSEs see a net increase. This means James-Stein cannot dominate the vanilla estimator.
- As we increase dimensions, the helped-hurt ratio is higher. By the time we have 3 or more dimensions, as we move X away from the origin, the helped-hurt ratio approaches 50-50 more slowly than the pair-wise weakness approaches 0. This means that its strength (the helped-hurt ratio) is able to overcome its weakness (pair-wise weakness) at every position of X, allowing James-Stein to dominate the vanilla estimator.

### Work Functions

```python
def calc_shrinkage_ran(shift, radii):
    pointcnt = 250
    X = np.random.normal(shift, radii, pointcnt)
    Y = np.random.normal(0, radii, pointcnt)

    before_sse = (X-shift)**2 + Y**2

    correction_factor = 1*radii**2/(X**2 + Y**2)/2
    afterX, afterY = (1-correction_factor)*X, (1-correction_factor)*Y

    after_sse = (afterX-shift)**2 + afterY**2

    coloring = np.array(['y'] * pointcnt) # about the same, default yellow
    coloring[after_sse - before_sse > 0.2] = 'r'    # deproved red
    coloring[after_sse - before_sse < -0.2] = 'b'   # improved blue
    return X, Y, afterX, afterY, coloring, before_sse, after_sse
```

```python
def calc_shrinkage_mesh(shift, radii):
    pointcnt = 26
    x = np.linspace(-radii, radii, pointcnt) + shift
    y = np.linspace(-radii, radii, pointcnt)
    X, Y = np.meshgrid(x, y)

    before_sse = (X-shift)**2 + Y**2

    correction_factor = 1/np.sqrt(X**2 + Y**2)
    afterX, afterY = (1-correction_factor)*X, (1-correction_factor)*Y

    after_sse = (afterX-shift)**2 + afterY**2

    coloring = np.array([['y'] * pointcnt] * pointcnt) # about the same, default yellow
    coloring[after_sse - before_sse > 0.1] = 'r'    # deproved red
    coloring[after_sse - before_sse < -0.1] = 'b'   # improved blue
    return X, Y, afterX, afterY, coloring, before_sse, after_sse
```

```python
def subplotter(X, Y, afterX, afterY, coloring, before_sse, after_sse, shift, ax):
    ax.scatter(X, Y, marker='.', c='grey', alpha=0.4, s=3)
    ax.scatter(afterX, afterY, marker='.', c=coloring.flatten(), s=4)
    ax.grid()
    ax.set_xlim(-5, 16)
    ax.set_ylim(-(2*radii+1), 2*radii+1)
    ax.text(x=shift-0.4, y=-0.4, s='X', size=12)
    ax.set_title(f'X centered at y={shift}\nTotal SSE Change: {int(sum(before_sse))} -> {int(sum(after_sse))}', size=12)
```

```python
def pairwise_compare(center):
    near, far = center - 1, center + 1
    near_factor, far_factor = 1/near**2, 1/far**2
    near_correction, far_correction = near_factor * near, far_factor * far
    print(f'Center: Pos {center:2}')
    print(f'  Near: Pos {  near:2}, Shrinkage {1 - near_factor:.3f}, Final Pos {near-near_correction:6.3f}, Delta {near_correction:.3f}')
    print(f'   Far: Pos {   far:2}, Shrinkage {1  - far_factor:.3f}, Final Pos {far-far_correction:6.3f}, Delta { far_correction:.3f}')
```
